```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np

def histogramme(url, divide=4):
    img=mpimg.imread(url)
    
    h = np.zeros((divide,divide,divide))
    sizeBlock = (256//divide)

    for row in img:
        for pixel in row:
            r = pixel[0]//sizeBlock
            g = pixel[1]//sizeBlock
            b = pixel[2]//sizeBlock
            h[r][g][b] += 1
    
    h = h/h.sum()
    return h

def chi2(x,y):
    x = np.ravel(x)
    y = np.ravel(y)
    dist = 0
    for xi in x:
        for yi in y:
            if(xi != 0 or yi != 0):
                dist += pow(xi-yi,2)/(xi+yi)
    return dist

h = histogramme('image2.jpg')
plt.hist(range(0,64),64,weights=np.ravel(h))

h2 = histogramme('image.jpg')

plt.hist(range(0,64),64,weights=np.ravel(h2))

chi2(h,h2)
        
```
